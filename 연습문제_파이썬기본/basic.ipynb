```python
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

for num in [5, 7]:
    print(f"{num}! (재귀) = {factorial_recursive(num)}")
    print(f"{num}! (반복) = {factorial_iterative(num)}")
```

```python
import math

def statistics(nums):
    avg = sum(nums)/len(nums)
    maximum = max(nums)
    minimum = min(nums)
    variance = sum((x-avg)**2 for x in nums)/ len(nums)
    std= math.sqrt(variance)
    return avg, maximum, minimum, std

data=[10,20,30,40,50]
avg, maxi,mini,std_val= statistics(data)
print(f"숫자들: {data}")
print(f"평균: {avg}")
print(f"최댓값: {maxi}")
print(f"최솟값: {mini}")
print(f"표준편차: {std_val:.2f}")
```
